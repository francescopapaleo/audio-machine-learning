# Normalize Data with sklearn

The formula for Normalization is given below:

$$
x_{norm} = {x - x_{min} \over x_{max} - x_{min}}
$$

We subtract the minimum value from each entry and then divide the result by the range. Where range is the difference between the maximum value and the minimum value.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from scipy.io import wavfile
```

```python
# initialize the input array
x_array = np.array([0, 0.9,2.3,3,3.9])

# call method to normalize from 0 to 1
norm_x = preprocessing.normalize([x_array])

#print output
print(norm_x)
```

```python
# Plot the original data
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x_array, marker='o', linestyle='-', color='b')
plt.title('Original Data')
plt.xlabel('Index')
plt.ylabel('Values')

# Plot the normalized data
plt.subplot(1, 2, 2)
plt.plot(norm_x.flatten(), marker='o', linestyle='-', color='r')
plt.title('Normalized Data (0 to 1)')
plt.xlabel('Index')
plt.ylabel('Normalized Values')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from sklearn import preprocessing

# Load an example audio file
sample_rate, audio_signal = wavfile.read('../data/audio/drum_loop_48kHz24bit.wav')

# Extract a portion of the audio signal for demonstration
start_time = 10000
end_time = 20000
subset_signal = audio_signal[start_time:end_time]

# Plot the original audio signal
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(subset_signal, color='b')
plt.title('Original Audio Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Normalize the audio signal
normalized_signal = preprocessing.normalize([subset_signal])

# Plot the normalized audio signal
plt.subplot(1, 2, 2)
plt.plot(normalized_signal.flatten(), color='r')
plt.title('Normalized Audio Signal')
plt.xlabel('Time')
plt.ylabel('Normalized Amplitude')

# Show the plots
plt.tight_layout()
plt.show()
```
